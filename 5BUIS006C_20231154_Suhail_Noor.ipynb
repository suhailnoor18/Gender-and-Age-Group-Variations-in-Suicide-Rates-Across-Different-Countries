```python

# Identify outliers using the IQR method
def identify_outliers(DataViz):
    Q1 = DataViz.quantile(0.25)
    Q3 = DataViz.quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((DataViz < (Q1 - 1.5 * IQR)) | (DataViz > (Q3 + 1.5 * IQR))).sum()
    return outliers

# Apply the function to the dataset
outliers = identify_outliers(DataViz.select_dtypes(include=['float64', 'int64']))

# Print the number of outliers in each variable
print("Number of outliers in each variable:")
print(outliers)

# Review the outliers by visualizing them using boxplots
def visualize_outliers(DataViz, column):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=DataViz, y=column)
    plt.title(f'Boxplot of {column}')
    plt.show()

# Visualize outliers for relevant columns
columns_to_review = ['suicides/100k pop', 'HDI for year', 'suicides_no']
for column in columns_to_review:
    visualize_outliers(DataViz, column)
```

```python
def remove_outliers(DataViz, columns):
    Q1 = DataViz[columns].quantile(0.25)
    Q3 = DataViz[columns].quantile(0.75)
    IQR = Q3 - Q1
    df_no_outliers = DataViz[~((DataViz[columns] < (Q1 - 1.5 * IQR)) | (DataViz[columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_no_outliers

# Columns to remove outliers from
columns_to_remove_outliers = ['suicides/100k pop', 'HDI for year', 'suicides_no']

# Print the number of data points before and after removing outliers
print(f"Number of data points before removing outliers: {len(DataViz)}")


# Remove outliers from the dataset
DataViz = remove_outliers(DataViz, columns_to_remove_outliers)

print(f"Number of data points after removing outliers: {len(DataViz)}")
```

```python
# Review the outliers by visualizing them using boxplots
def visualize_outliers(DataViz, column):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=DataViz, y=column)
    plt.title(f'Boxplot of {column}')
    plt.show()

# Visualize outliers for relevant columns
columns_to_review = ['suicides/100k pop', 'HDI for year', 'suicides_no']
for column in columns_to_review:
    visualize_outliers(DataViz, column)
```

Exploratory Data Analysis

```python
# Load the dataset
cleanedset = pd.read_csv('/content/drive/MyDrive/Data Viz CW/cleanedset.csv')
```

Univariate Analysis

```python
# Distribution of Suicides by Gender (Pie Chart)
gender_counts = cleanedset['sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Suicides by Gender')
plt.show()
```

```python
#Distribution of Suicides by Age Group (Histogram)
plt.figure(figsize=(10, 6))
plt.hist(cleanedset['mean_age'], bins=range(0, 100, 10), edgecolor='black')
plt.title('Distribution of Suicides by Age Group')
plt.xlabel('Mean Age')
plt.ylabel('Frequency')
plt.show()
```

Multivariate Analysis

```python
#Suicides by Gender and Age Group (Box Plot)
plt.figure(figsize=(12, 8))
sns.boxplot(data=cleanedset, x='mean_age', y='suicides/100k pop', hue='sex')
plt.title('Suicides by Gender and Age Group')
plt.xlabel('Mean Age')
plt.ylabel('Suicides per 100k Population')
plt.legend(title='Gender')
plt.show()
```

```python
#Suicides by Gender and Country (Stacked Bar Chart)
country_gender_counts = cleanedset.groupby(['country', 'sex']).size().unstack()
country_gender_counts.plot(kind='bar', stacked=True, figsize=(14, 10))
plt.title('Suicides by Gender and Country')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.legend(title='Gender')
plt.show()
```

```python
#Suicides by Age Group and Country (Heatmap)
age_country_counts = cleanedset.pivot_table(index='country', columns='mean_age', values='suicides_no', aggfunc='sum')
plt.figure(figsize=(14, 10))
sns.heatmap(age_country_counts, cmap='YlGnBu', linewidths=.5)
plt.title('Suicides by Age Group and Country')
plt.xlabel('Mean Age')
plt.ylabel('Country')
plt.show()
```

```python
# Bivariate analysis: Relationship between GDP per Capita and Population Size (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=cleanedset, x='gdp_per_capita ', y='population', hue='sex')
plt.title('Relationship between GDP per Capita and Population Size')
plt.xlabel('GDP per Capita')
plt.ylabel('Population Size')
plt.legend(title='Gender')

# Add trend line:
sns.regplot(data=cleanedset, x='gdp_per_capita ', y='population', scatter=False, color='black')
plt.show()
```
